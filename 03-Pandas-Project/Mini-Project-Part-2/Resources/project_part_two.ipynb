```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = ("Resources/2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
df = pd.read_csv(file, encoding="ISO-8859-1", low_memory= False)
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
new_df = df.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_df.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
new_df = new_df.replace({0.0: 'No', 1.0: 'Yes'})
new_df.head()
```

```python
# Extract rows for only those who attended a bootcamp
attended_bootcamp = new_df.loc[new_df['AttendedBootcamp'] == 'Yes']
attended_bootcamp.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamp_name = pd.DataFrame(df["BootcampName"].value_counts().rename_axis("BootcampName").reset_index(name = "Attendees"))
bootcamp_name
```

```python
attended_bootcamp = new_df.loc[new_df['AttendedBootcamp'] == 'Yes']
attended_bootcamp = attended_bootcamp.loc[new_df['BootcampRecommend'] == 'Yes']

# Count how many attendees of each bootcamp would recommend the bootcamp
sum_recomended = pd.DataFrame(attended_bootcamp.groupby('BootcampName')["BootcampRecommend"].count())
sum_recomended.reset_index(inplace=True)
sum_recomended.columns=["BootcampName", "Recommenders"]

sum_recomended
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged_df = pd.merge(sum_recomended, bootcamp_name, on="BootcampName")
merged_df.
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merged_df["% Recommend"] = round(merged_df["Recommenders"]/ merged_df["Attendees"]*100,2).map("{0:,.2f}%".format)
merged_df = merged_df.sort_values(["% Recommend"], ascending=False)
merged_df.head()
```

```python
# Export to excel and remove index
merged_df.to_excel("output/Bootcamp2.xlsx", index=False)
```
